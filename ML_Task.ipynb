Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Data Extraction

```python
TrainData = pd.read_csv("/content/drive/MyDrive/train_updated.csv")
TestData = pd.read_csv("/content/drive/MyDrive/test.csv")
trainData = TrainData.to_numpy()
testData = TestData.to_numpy()
x_train = trainData[:,0]
y_train = trainData[:,1]
x_test = testData[:,0]
y_test = testData[:,1]
```

Model Training

```python
def Linear_reg(alpha,no_of_iterations):
  w = 0 #parameter
  b = 0 #bias
  m = len(y_train)
  for i in range(no_of_iterations):
    y_hat = x_train * w + b #hypothesis calculation
    j = np.sum(np.square(y_train - y_hat))/(2*m)  # Cost function calculation
    dj_dw = -np.sum(np.dot(x_train.T,(y_train - y_hat)))/m
    dj_db = -np.sum(y_train - y_hat)/m
    w = w - alpha * dj_dw # parameter updation
    b = b - alpha * dj_db # bias updation
    if i%1000==0:
      print(j)
  #   plt.scatter(i,j)
  # plt.show()
  plt.title("Training")
  plt.plot(x_train,y_hat,color = 'r', label = "predicted")
  plt.scatter(x_train,y_train, label = "Train data")
  plt.legend()
  plt.show()
  return w,b
```

Model Testing

```python
W,B = Linear_reg(5.9e-4,10000)
y_predicted = x_test * W + B # Prediction of data by model
plt.title("Testing")
plt.scatter(x_test,y_test,label = "Test Data")
plt.plot(x_test,y_predicted,color = 'r', label = "predicted")
plt.legend()
plt.show()
n = len(y_test)
accuracy = np.sqrt(np.sum(np.square(y_test - y_predicted))/(2*n)) #Accuracy of predicted values
print(accuracy)
```
